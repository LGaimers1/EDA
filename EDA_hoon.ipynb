```python
def get_count_0(series):
    res = len([ _ for x in series if x==0])
    return res
```

```python
def get_percentage_0(series):
    res = len([ _ for x in series if x==0])
    cnt = len([ _ for x in series])
    per = res/cnt
    return per
```

```python
def get_percentage_1(series):
    res = len([ _ for x in series if x==1])
    cnt = len([ _ for x in series])
    per = res/cnt
    return per
```

```python
def get_percentage_2(series):
    res = len([ _ for x in series if x==2])
    cnt = len([ _ for x in series])
    per = res/cnt
    return per
```

```python
train_data.groupby('LINE').agg({'Y_Class':get_percentage_0}).reset_index()
```
